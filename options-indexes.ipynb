```python
import os.path
import pandas
```

```python
all_data = None

for day in range(1, 31):
    filename = f'options-data/bb_options_201903{day:02d}.csv'
    
    if os.path.isfile(filename):
        day_data = pandas.read_csv(filename)
        day_data['QuoteDate'] = f'03/{day:02d}/2019'

        if all_data is None:
            all_data = day_data
        else:
            all_data = all_data.append(day_data)

all_data
```
